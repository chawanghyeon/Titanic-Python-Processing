```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn')
sns.set(font_scale=2.5)
import missingno as msno

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
df_train = pd.read_csv('../input/train.csv')
df_test = pd.read_csv('../input/test.csv')
```

```python
for col in df_train.columns:
    msg = 'column: {:>11}\t Percent of NaN value: {:.2f}%'.format(col, 100 * (df_train[col].isnull().sum() / df_train[col].shape[0]))
    print(msg)
```

```python
for col in df_test.columns:
    msg = 'column: {:>11}\t Percent of NaN value: {:.2f}%'.format(col, 100 * (df_test[col].isnull().sum() / df_test[col].shape[0]))
    print(msg)
```

```python
# 
msno.matrix(df_train.iloc[:, :], figsize=(8, 8), color=(0.8, 0.5, 0.2))
```

```python
msno.matrix(df_test.iloc[:, :], figsize=(8, 8), color=(0.8, 0.5, 0.2))
```

```python
msno.bar(df=df_train.iloc[:, :], figsize=(8, 8), color=(0.8, 0.5, 0.2))
```

```python
msno.bar(df=df_test.iloc[:, :], figsize=(8, 8), color=(0.8, 0.5, 0.2))
```

```python
f, ax = plt.subplots(1, 2, figsize=(18, 8))

df_train['Survived'].value_counts().plot.pie(explode=[0, 0.1], autopct='%1.2f%%', ax=ax[1], shadow=True)
ax[1].set_title('Pie plot - Survived')
ax[1].set_ylabel('')
sns.countplot('Survived', data=df_train, ax=ax[0])
ax[0].set_title('Count plot - Survived')

plt.show()
```

```python
df_train[['Pclass', 'Survived']].groupby(['Pclass'], as_index=True).count().reset_index()
```

```python
df_train[['Pclass', 'Survived']].groupby(['Pclass'], as_index=True).sum().reset_index()
```

```python
df_train[['Pclass', 'Survived']].groupby(['Pclass'], as_index=True).mean().sort_values(by='Survived', ascending=False).plot.bar(figsize=(8, 8))
```

```python
f, ax = plt.subplots(1, 2, figsize=(18, 8))

df_train['Pclass'].value_counts().plot.bar(color=['#CD7F32','#FFDF00','#D3D3D3'], ax=ax[0])
ax[0].set_title('Number of Passengers By Pclass')
ax[0].set_ylabel('Count')
sns.countplot('Pclass', hue='Survived', data=df_train, ax=ax[1])
ax[1].set_title('Pclass: Survived vs Dead')

plt.show()
```

```python
f, ax = plt.subplots(1, 2, figsize=(18, 8))

df_train[['Sex', 'Survived']].groupby(['Sex']).mean().plot.bar(ax=ax[0])
ax[0].set_title('Survived vs Sex')
sns.countplot('Sex', hue='Survived', data=df_train, ax=ax[1]) # hue   
ax[1].set_title('Sex: Survived vs Dead')

plt.show()
```

```python
df_train[['Sex', 'Survived']].groupby(['Sex'], as_index=False).mean().sort_values(by='Survived')
```

```python
df_train[['Sex', 'Survived']].groupby(['Sex'], as_index=False).sum().sort_values(by='Survived')
```

```python
#       
df_train[['Sex', 'Survived']].groupby(['Sex'], as_index=False).count().sort_values(by='Survived')
```

```python
pd.crosstab(df_train['Sex'], df_train['Survived'], margins=True)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(9, 5))

sns.kdeplot(df_train[df_train['Survived'] == 1]['Age'], ax=ax) #   
sns.kdeplot(df_train[df_train['Survived'] == 0]['Age'], ax=ax)
plt.legend(['Survived == 1', 'Survived == 0'])

plt.show()
```
