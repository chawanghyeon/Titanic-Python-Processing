```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("input/avg15.09.csv", thousands=',', encoding = 'utf-8')
df1 = df.iloc[3:]
df1.columns = list(df.iloc[2])
col = list(df.iloc[2])
for i in range(2, len(col)-3):
    if i == 2:
        col[2] = '2013년 12월'
    elif 3<=i<=14:
        col[i] = '2014년 ' + col[i]
    else: 
        col[i] = '2015년 ' + col[i]
df1.columns = col
lis1 = list(df1.iloc[:, 0])
lis2 = list(df1.iloc[:, 1])
check = list(df1.iloc[:, 1].isnull())
check[0]
for i in range(len(lis2)):
    if check[i] == True:
         lis2[i] = lis1[i]

df1['시군구'] = lis2
df1 = df1.drop(['시도'], axis = 1)
df1 = df1.dropna(axis = 0)
col = list(df.iloc[2])
```

```python
df2 = pd.read_csv('input/avg18.07.csv', thousands=',')
df2.head(12)
```

```python
df3 = df2.loc[df2['규모구분']=='전체']
df3.head(5)
```

```python
df3.drop(columns=['규모구분'])
df3 = df3.sort_values(by=['연도', '월'])
df3
```

```python
df3['연도월'] = df3['연도'].astype(str) + '년 ' + df3['월'].astype(str) + '월'
col2 = df3['연도월']
df3
```

```python
df3['분양가격(㎡)'] = df3['분양가격(㎡)'].str.replace(',', '')
df3['분양가격(㎡)'] = pd.to_numeric(df3['분양가격(㎡)'], errors='coerce')
# df3 = df3.dropna(data=['분양가격(㎡)'])
df3['분양가격(㎡)'] = df3['분양가격(㎡)'].astype(int)
```

```python
for i in df1.loc[:, '2013년 12월':]:
    df1[i] = df1[i].str.replace(',', '')
    df1[i] = pd.to_numeric(df1[i], errors='coerce')
    # df3 = df3.dropna(data=['분양가격(㎡)'])
    df1[i] = df1[i].astype(int)
```

```python
temp = pd.DataFrame(index=df3['연도월'].unique(), columns=[df3['지역명'].unique()])
temp = temp.transpose()

for i in df3[['연도월', '지역명', '분양가격(㎡)']].values:
    temp[i[0]][i[1]] = int(int(i[2])*3.3)

temp.reset_index(inplace=True)
temp.sort_values(by=['level_0'], axis=0, inplace=True)
del temp['level_0']
```

```python
df1.drop(3, inplace=True)
df1.drop(13, inplace=True)
df1.reset_index(inplace=True)
del df1['index']
del df1['전월비']
del df1['전년말비']
del df1['전년동월비']
df1.sort_values(by=['시군구'], axis=0, inplace=True)
```

```python
result = pd.concat([df1, temp], axis=1)
result.set_index('시군구', inplace=True)
result
```

```python
# 제주도 평균 분양가(1평당)
result.loc['제주'].mean()
```
